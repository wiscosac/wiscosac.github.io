```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.maxpreps.com/nj/football/stat-leaders/offense/passing/yds/"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'lxml') # or 'html.parser'
```

```python
headers = [a.get_text(strip=False) for a in soup.find_all('th')]
headers
```

```python
dataframe = []
rows = [a for a in soup.find_all('tr')]

for row in rows:
    soup = BeautifulSoup(str(row), "html.parser")
    
    data = [td.get_text(strip=False) for td in soup.find_all(['td','span','a'], recursive=True)]
    if len(data)==1:continue
    del data[1] # removes the name, position, school merged cell
    dataframe += [data]
    print(data)
```

```python
# headers don't have position and school as they are in the name column so for this example I had to define it manually
columns = ['Rank', 'Name', 'Position', 'School', 'Yards', 'Avg', 'Comp', 'Att', 'Pct', 'TD', 'Int', 'Long', 'Games']
df = pd.DataFrame(dataframe, columns = columns)
```
